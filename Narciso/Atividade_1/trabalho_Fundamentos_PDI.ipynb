```python
#criar variáveis para coleta das informações de pixels de maior e menor intensidade
maxR=None
minR=None
maxG=None
Ming=None
maxB=None
minB=None
```

```python
#coletar dimensões da imagem
linhas,colunas,dimensao=imagem.shape
```

```python
for i in range(linhas):
    for k in range(colunas):
        #coletar maior pixel camara red
        if((i==0 and k==0) or maxR<imagem[i,k,0]):
            maxR=imagem[i,k,0]
        #coletar menor pixel camara red
        if((i==0 and k==0) or minR>imagem[i,k,0]):
            minR=imagem[i,k,0]
            
        #coletar maior pixel camara green
        if((i==0 and k==0) or maxG<imagem[i,k,1]):
            maxG=imagem[i,k,1]
        #coletar menor pixel camara green
        if((i==0 and k==0) or minG>imagem[i,k,1]):
            minG=imagem[i,k,1]
            
        #coletar maior pixel camara blue
        if((i==0 and k==0) or maxB<imagem[i,k,2]):
            maxB=imagem[i,k,2]
        #coletar menor pixel camara blue
        if((i==0 and k==0) or minB>imagem[i,k,2]):
            minB=imagem[i,k,2]
```

```python
print("Max Red: {}\nMax Green: {}\nMax Blue: {}".format(maxR,maxG,maxB))
print("Min Red: {}\nMin Green: {}\nMin Blue: {}".format(minR,minG,minB))
```

4) Leia uma imagem colorida, calcule as componentes no modelo de cores RGB.  Atribua o valor 255 a todos os pixels com intensidade maior que 150, faça isso em cada componente. Após isso, junte as componentes para formar uma imagem colorida novamente e salve o resultado.

```python
#importar a funcao imread da biblioteca skimage
from skimage.io import imread

#Lendo a imagem colorida
imagem = imread('gatos.png')
```

```python
#coletando as camadas separadamente
red = imagem[:,:,0]
green = imagem[:,:,1]
blue = imagem[:,:,2]
```

```python
#coletando as dimensões da imagem
linhas,colunas=red.shape
```

```python
#percorrer toda a imagem e corrigir pixels 150 para 255(em cada camada)
for i in range(linhas):
    for k in range(colunas):
        if(red[i,k] > 150):
            red[i,k]=255
        if(green[i,k] > 150):
            green[i,k]=255
        if(blue[i,k] > 150):
            blue[i,k]=255
```

```python
#importar função numpy
import numpy as np
```

```python
#juntar as camadas ajustadas
imagem_ajustada=[]
imagem_ajustada = np.array(np.dstack((red,green,blue)), dtype=np.uint8)
```

```python
#importar função imsave da biblioteca skimage
from skimage.io import imsave

#salvar imagem em níveis de cinza
imsave('gatos_cinza_acima150to255.png',imagem_ajustada)
```

8) Recorte uma imagem da seguinte forma:
- O corte será feito no canal de cor com a maior media de pixels.
- O centro do corte deve ser o pixel com o maior desvio padrão em uma janela 3x3.
- O tamanho do recorte será 1/4 do tamanho da imagem. Ex: Imagem 100 x 80 -> Recorte 25 x 20
- Cuidado com as bordas! Caso uma janela fique fora da imagem, o centro do recorte deverá ser alterado.

```python
#importar a funcao imread da biblioteca skimage
from skimage.io import imread

#importar função skimage para conversão em níveis de cinza
from skimage.color import rgb2gray

#importar função numpy
import numpy as np

#importar a biblioteca matplotlib
import matplotlib.pyplot as plt
%matplotlib inline


#Lendo a imagem colorida
imagem = imread('gatos.png')
```

```python
#coletar as médias de cada camada
med_red = np.mean(imagem[:,:,0])
med_green = np.mean(imagem[:,:,1])
med_blue = np.mean(imagem[:,:,2])
linhas,colunas,dimensao = imagem.shape
```

```python
x=0
y=0
dp=None

#coletar posição maior desvio padrão de acordo com a camada de maior média
if (med_red>=med_green and med_red >= med_blue):
    for l in range(linhas-3):
        for c in range(colunas-3):
            k=imagem[l:(l+3),c:(c+3),0]
            #desvio_padrao=np.std(k)
            if(dp==None or np.std(k)>dp):
                x=l
                y=c
                dp=np.std(k)
elif(med_green >= med_red and med_green>=med_blue):
    for l in range(linhas-3):
        for c in range(colunas-3):
            k=imagem[l:(l+3),c:(c+3),1]
            #desvio_padrao=np.std(k)
            if(dp==None or np.std(k)>dp):
                x=l
                y=c
                dp=np.std(k)
elif(med_blue >= med_red and med_blue>=med_green):
    for l in range(linhas-3):
        for c in range(colunas-3):
            k=imagem[l:(l+3),c:(c+3),2]
            #desvio_padrao=np.std(k)
            if(dp==None or np.std(k)>dp):
                x=l
                y=c
                dp=np.std(k)
                
```

```python
#coletar o tamanho do recorte (1/4 da imagem), convertendo-os em inteiro
l_corte=int(linhas/4)
c_corte=int(colunas/4)
```

```python
#validar se o xMax não estrapola o tamanho da imagem
xMax=(x+1)+(l_corte/2)
if(xMax>linhas):
    x=int(linhas-(l_corte/2))
```

```python
#validar se o yMax não estrapola o tamanho da imagem
yMax=(y+1)+(c_corte/2)
if(yMax>colunas):
    y=int(colunas-(c_corte/2))
```

```python
xMin=(x+1)-(l_corte/2)
if(xMin<0):
    x=int(linhas+(l_corte/2))
```

```python
yMin=(y+1)-(c_corte/2)
if(yMin<0):
    y=int(colunas+(c_corte/2))
```

```python
#colocar em variáveis os valores do recorte para facilitar
xMin=int(x-l_corte/2)
xMax=int(x+l_corte/2)
yMin=int(y-c_corte/2)
yMax=int(y+c_corte/2)
```
